```python
import statsmodels.api as sm
import pandas as pd
from patsy import dmatrices
```

```python
#dataset-house price prediction dataset
df = pd.read_csv('train.csv')
cars=['MSSubClass','OverallQual','OverallCond','YearBuilt','YearRemodAdd','TotalBsmtSF','SalePrice']
df=df[cars]
df[-5:]
df.tail(5)
df
```

```python
# count null values
df.isnull().sum()
df.tail(5)
```

```python
#Statsmodel
y, X = dmatrices('SalePrice ~ OverallQual + OverallCond + YearBuilt + YearRemodAdd + TotalBsmtSF', data=df, return_type='dataframe')
```

```python
print(y)
print(X)
```

```python
#model
model=sm.OLS(y,X)
res=model.fit()
print(res.summary())
```

```python
res.params
```
