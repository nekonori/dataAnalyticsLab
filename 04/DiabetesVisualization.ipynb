```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
sns.countplot(x=df['Outcome'])
plt.show()
```

```python
labels=['No diabetes','Diabetes']
plt.pie(df['Outcome'].value_counts(),labels=labels,autopct='%1.1f%%')
```

```python
sns.boxplot(x='BMI',data=df)
```

```python
df.plot(kind ="scatter",
x ='Outcome',
y ='Pregnancies')
plt.plot()
```

```python
sns.heatmap(df.corr(),annot=True)
```

###Line Plot

```python
newdf=df.loc[1:50]
var=np.array(newdf['BMI'])
plt.plot(var,linestyle='dotted')
```

###Pair Plot

```python
newdf=newdf[['Glucose','Pregnancies']]
sns.pairplot(data=newdf,kind='scatter')
```

###Viloin Plot

```python
sns.violinplot(x='Outcome',y='Pregnancies',data=df,size=6)
```

###Histogram

```python
sns.histplot(data=df,x='Outcome')
```
