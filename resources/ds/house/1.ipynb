## Univariate analysis

```python
# Set the color palette
colors = sns.color_palette('muted')

# Create the count plot
sns.countplot(x='grade of the house', data=df, palette=colors)

unique_values = sorted(df['grade of the house'])

# Add annotations to the bars
for i in df['grade of the house'].unique():
    count = df[df['grade of the house'] == i].shape[0]
    plt.text(i-4, count+50, count, ha='center', fontsize=12)

# Set the title and axis labels
plt.title('Count Plot of Grade of the House', fontsize=16)
plt.xlabel('Grade of the House', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.figure(figsize=(10, 6))
# Show the plot
plt.show()
```

```python
sns.histplot(x='Area of the house(excluding basement)', data=df)

# Set plot title and labels
plt.title('Histogram of Area of the house(excluding basement)')
plt.xlabel('Area of the house(excluding basement)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
sns.boxplot(data=df, x='grade of the house')
plt.title('Box plot of Grade of the house')
plt.xlabel('Grade of the house')
plt.ylabel('Frequency')
```

```python
sns.violinplot(x='living area', data=df)
plt.title('Violin plot of the living area')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(data=df, x='number of floors', color='blue', shade=True)
plt.title('Kernel density estimate of the number of floors')
```

## Bivariate analysis

```python
sns.pairplot(df[['living area', 'lot area', 'Price']])
plt.suptitle('Pairplot of Living Area, Lot Area and Price')
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
# Plot violin plot of 'grade of the house' vs 'price'
sns.violinplot(x='grade of the house', y='Price', data=df)
plt.title('Violin Plot of Grade of the House vs Price')
plt.xlabel('Grade of the House')
# Author Pragadesh BS
plt.ylabel('Price')
plt.show()
```

```python
# Plot scatter plot of 'living area' vs 'price'
sns.scatterplot(x='living area', y='Price', data=df)
plt.title('Scatter Plot of Living Area vs Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
corr = df[['Area of the house(excluding basement)', 'Price']].corr()

# Heatmap of correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Box plot of 'number of bedrooms' vs 'Price'
sns.boxplot(x='number of bedrooms', y='Price', data=df)
plt.title('Box Plot of Number of Bedrooms vs Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```
