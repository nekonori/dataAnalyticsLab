# Dealing with missing values

```python
if df.isnull().values.any():
    print("The dataframe contains null values.")
else:
    print("The dataframe does not contain null values.")
```

# Encoding categorical values

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print(cat_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cat_cols:
  df[col] = le.fit_transform(df[col])

df.tail()
```

# Outliers

```python
print("shape before removing outliers", df.shape)

from scipy import stats

# Calculate z-scores for all columns
for col in df.columns:
  z_scores = stats.zscore(df[col])

  # Define threshold value
  threshold = 3

  # Remove data points with z-score greater than threshold
  df = df[(z_scores < threshold)]

print("shape after removing outliers", df.shape)
```

# Data Split

```python
# X -> Independant variable, y -> dependant variable

X = df.drop('price', axis=1)
y = df['price']
```

# Scaling independant variables

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df
```

# Split into training and testing dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Building the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

# Training the model

```python
lr.fit(X_train, y_train)
```

# Test the model

```python
lr.predict(X_test.iloc[:5, :])
```

# Measuring the performance using Metrics

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the testing set
y_pred = lr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print(f"Mean squared error: {mse:.2f}")
print(f"Root mean squared error: {rmse:.2f}")
print(f"R-squared: {r2:.2f}")
```
